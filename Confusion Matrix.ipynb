```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score

# Simulated ground truth and predictions
# TP = 180, FN = 50, FP = 70, TN = 500
y_true = [1]*180 + [1]*50 + [0]*70 + [0]*500  # 180 TP, 50 FN, 70 FP, 500 TN
y_pred = [1]*180 + [0]*50 + [1]*70 + [0]*500

# Generate confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=[1, 0])

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Readmit", "No Readmit"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix – Hospital Readmission Prediction")
plt.tight_layout()

# Calculate precision and recall
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)

precision, recall, cm
```
